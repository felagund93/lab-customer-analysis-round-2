```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
#Import the data files into pandas DataFrame
df = pd.read_csv('files_for_lab/csv_files/marketing_customer_analysis.csv')

print('Shape of dataframe is (rows, columns):', np.shape(df))
    
display(df)
```

```python
df.info()
```

```python
##Tasks##

#drop Unnamed
#Check gender
#Check state
#Round Costumer Lifetime Value, Total Claim Amount
#Income to float
#Number of open complaints to int
```

```python
#Standardisation of column names
df.columns = [df.columns[i].lower().replace(' ','_') for i in range(len(df.columns))]

display(df.columns)
```

```python
df.info()
display(df.isna().sum())
df['vehicle_type'].unique() # check the unique values in the column
```

```python
#Decided to drop the columns Unnamed: 0 and vehicle_type. 
#They don't contribute to anything relevant in the dataframe

cols_to_drop=['unnamed:_0','vehicle_type']
df=df.drop(cols_to_drop, axis=1)
df.info()
```

```python
display(df['vehicle_class'].unique())
display(df['vehicle_size'].unique())
```

```python
df['vehicle_class'].fillna('Other', inplace=True)
df['vehicle_class'].isna().sum()
display(df['vehicle_class'].unique())

df['vehicle_size'].fillna('n/a', inplace=True)
df['vehicle_size'].isna().sum()
display(df['vehicle_size'].unique())
```

```python
df.info()
```

```python
#Fill in the empty "months_since_last_claim" with the average value, rounded to int, and transform everything 
#to int

avg_months_since_last_claim=int(df['months_since_last_claim'].dropna().mean())
df['months_since_last_claim']=df['months_since_last_claim'].fillna(avg_months_since_last_claim).astype('int64')
```

```python
#Fill in the empty "number_of_open_complaints" with the average value, rounded to int, and transform everything 
#to int

display(df['number_of_open_complaints'])
df['number_of_open_complaints'].hist()
plt.show()

#The average open complaints is 0. The nan's are filled with 0.

df['number_of_open_complaints']=df['number_of_open_complaints'].fillna(0).astype('int64')
df['number_of_open_complaints'].hist()
plt.show()
```

```python
df.info()
df.head()
```

```python
display(df['state'].unique())
display(df['response'].unique())
df['state']=df['state'].fillna('Other')
df['response']=df['response'].fillna('n/a')

display(df['state'].unique())
display(df['response'].unique())
```

```python
#Round Costumer Lifetime Value, Total Claim Amount
cols_to_round=['customer_lifetime_value','total_claim_amount']
for i in cols_to_round:
    df[i]=df[i].round(decimals=2)
#Income to float
df['income']=df['income'].astype('float64').round(decimals=3)
```
