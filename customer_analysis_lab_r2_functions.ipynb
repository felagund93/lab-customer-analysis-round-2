```python
#Importing DF, saving it as global, and returning shape
def imp_and_shape(path):
    import pandas as pd
    import numpy as np
    global df
    df = pd.read_csv(path)
    return 'Shape of dataframe is (rows, columns):', np.shape(df)

#Standardisation of column names
def col_names_std(df):
    import pandas as pd
    import numpy as np
    df.columns = [df.columns[i].lower().replace(' ','_') for i in range(len(df.columns))]
    return df

#Drop columns from a DF using a list of columns
def col_drop(df, col):
    import pandas as pd
    import numpy as np
    df=df.drop(col, axis=1)
    return df

#Fill null values in a column from a DF (inplace) with a desired value
def fill_val(df, col, val):
    import pandas as pd
    import numpy as np
    df[col].fillna(val, inplace=True)
    return df

#Fill in an empty column with the average value, rounded to int, and transform everything to int
def fill_mean(df, col):
    import pandas as pd
    import numpy as np
    df[col]=df[col].fillna(int(df[col].dropna().mean())).astype('int64')
    return df
    
#Create a new column with "months", using a datetime object
def get_month(df, col):
    import pandas as pd
    import numpy as np
    import datetime
    from datetime import date
    df[col] = pd.to_datetime(df[col], errors='coerce')
    df['months']=pd.DatetimeIndex(df[col]).month
    return df
```
